```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# creating a sample dataset
np.random.seed(42)
n = 1000
age = np.random.normal(35, 10, n)
income = np.exp(np.random.normal(10, 0.5, n))
education_level = np.random.choice(['High School', 'Bachelor', 'Master', 'PhD'], n)
marital_status = np.random.choice(['Single', 'Married', 'Divorced'], n)
gender = np.random.choice(['Male', 'Female', 'Other'], n, p=[0.49, 0.49, 0.02])
location = np.random.choice(['Downtown', 'Suburb', 'Rural'], n)
```

```python
# creating a dataframe instance from the above dataset
df = pd.DataFrame({
    'Age': age,
    'Income': income,
    'Education': education_level,
    'Marital_Status': marital_status,
    'Gender': gender,
    'Location': location
})

"""
OR
df = pd.read_csv('population.csv')
"""
```

```python
# histogram: Age distribution
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=20, edgecolor='black')
plt.title('distribution of Ages in the population')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# bar plot: marital status
plt.figure(figsize=(10, 8))
df['Marital_Status'].value_counts().plot(kind='bar')
plt.title('Distribution of Marital Status')
plt.xlabel('Marital status')
plt.ylabel('Count')
plt.show()
```

```python
# pie chart: Gender
plt.figure(figsize=(10, 8))
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Gender distribution')
plt.ylabel('')
plt.show()
```

```python
# box plot: Income
plt.figure(figsize=(10, 8))
sns.boxplot(x=df['Income'])
plt.title('Distribution of Income levels')
plt.xlabel('Income')
plt.show()
```

```python
# scatter plot: Education level vs Income
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Education', y='Income', data=df)
plt.title('Education level vs Income')
plt.xlabel('Education level')
plt.ylabel('Income')
plt.show()
```

```python
# line plot: Average age over time
years = range(2010, 2025)
avg_age = np.random.normal(35, 2, len(years))

plt.figure(figsize=(10, 8))
plt.plot(years, avg_age, marker='o')
plt.title('Average age over time')
plt.xlabel('Year')
plt.ylabel('Average age')
plt.show()
```

```python
# violin plot: Education level
plt.figure(figsize=(10, 8))
sns.violinplot(x='Education', y='Age', data=df)
plt.title('Distribution of Age by Education level')
plt.xlabel('Education level')
plt.ylabel('Age')
plt.show()
```

```python
# heatmap: Education level by Marital status
plt.figure(figsize=(10, 8))
cross_tab = pd.crosstab(df['Education'], df['Marital_Status'])
sns.heatmap(cross_tab, annot=True, cmap='YlGnBu')
plt.title('Education level vs Marital status')
plt.xlabel('Marital status')
plt.ylabel('Education level')
plt.show()
```

```python
# area plot: location over time (simulated)
years = range(2010, 2025)
downtown = np.random.normal(40, 5, len(years))
suburb = np.random.normal(35, 3, len(years))
rural = np.random.normal(25, 2, len(years))

plt.figure(figsize=(10, 8))
plt.stackplot(years, downtown, suburb, rural, labels=['Downtown', 'Suburb', 'Rural'])
plt.title('Population density by Location over time')
plt.xlabel('year')
plt.ylabel('population')
plt.legend(loc='upper left')
plt.show()
```
