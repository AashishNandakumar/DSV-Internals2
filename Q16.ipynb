```python
import matplotlib.pyplot as plt
```

```python
# data for creating the pie chart
languages = ['Hindi', 'Bengali', 'Marathi', 'Telugu', 'Tamil', 'Gujarati', 'Urdu', 'Kannada', 'Odia', 'Malayalam']
speakers = [44.0, 8.1, 7.2, 6.9, 6.3, 5.0, 4.5, 3.8, 3.2, 3.2]
```

```python
# create the pie chart
plt.figure(figsize=(20, 8))  # create a figure of the required size

plt.pie(speakers, labels=languages, autopct='%1.1f%%')
plt.title('Distribution of Speakers for Major Indian regional languages')
plt.legend(title='Languages', loc='center right', bbox_to_anchor=(1, 0.5))

plt.tight_layout()
plt.show()
```
